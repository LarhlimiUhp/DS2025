```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.patches import Rectangle

# Configuration du style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# ============================================
# 1. DONN√âES PIB 2024
# ============================================

data = {
    'Pays': ['√âtats-Unis', 'Chine', 'Allemagne', 'Japon', 'Inde',
             'Royaume-Uni', 'France', 'Italie', 'Russie', 'Br√©sil',
             'Canada', 'Cor√©e du Sud', 'Espagne', 'Australie', 'Mexique',
             'Indon√©sie', 'Pays-Bas', 'Arabie Saoudite', 'Turquie', 'Pologne'],
    'PIB_Mds_USD': [29167.78, 18273.36, 4710.03, 4070.09, 3900,
                     3500, 3200, 2350, 2300, 2188.42,
                     2150, 1850, 1650, 1752.19, 1550,
                     1450, 1100, 1237.53, 1150, 850],
    'Croissance_%': [2.1, 3.0, 1.8, -0.5, 7.0,
                      4.1, 2.6, 1.5, 3.5, 2.8,
                      1.9, 2.4, 5.5, 2.2, 3.1,
                      5.0, 1.6, 2.7, 4.5, 4.9],
    'PIB_Habitant_USD': [86530, 12970, 56200, 32380, 2730,
                          51700, 47500, 39800, 15850, 10300,
                          55300, 35800, 34700, 67500, 11900,
                          5200, 63000, 34100, 13400, 22400],
    'Region': ['Am√©rique du Nord', 'Asie', 'Europe', 'Asie', 'Asie',
               'Europe', 'Europe', 'Europe', 'Europe/Asie', 'Am√©rique du Sud',
               'Am√©rique du Nord', 'Asie', 'Europe', 'Oc√©anie', 'Am√©rique du Nord',
               'Asie', 'Europe', 'Moyen-Orient', 'Europe/Asie', 'Europe']
}

df = pd.DataFrame(data)
df['Rang'] = range(1, len(df) + 1)

# Donn√©es d'√©volution USA vs Chine vs UE
evolution_data = {
    'Annee': [2020, 2021, 2022, 2023, 2024],
    'USA': [20936, 23315, 25464, 27361, 29168],
    'Chine': [14723, 17734, 17963, 17786, 18273],
    'UE': [15189, 17176, 16739, 18351, 17510]
}
df_evolution = pd.DataFrame(evolution_data)

# Donn√©es par r√©gion
region_data = {
    'Region': ['Am√©rique du Nord', 'Asie', 'Europe', 'Am√©rique du Sud', 'Moyen-Orient', 'Oc√©anie'],
    'PIB_Mds_USD': [32867.78, 30593.45, 22360.03, 2188.42, 1237.53, 1752.19],
    'Pourcentage': [32.5, 30.2, 22.1, 2.2, 1.2, 1.7]
}
df_regions = pd.DataFrame(region_data)

# ============================================
# 2. AFFICHAGE DES STATISTIQUES
# ============================================

print("=" * 80)
print(" üìä ANALYSE DU PIB MONDIAL 2024 - TOP 20 ".center(80, "="))
print("=" * 80)
print()

# Statistiques globales
pib_total = df['PIB_Mds_USD'].sum()
croissance_moyenne = df['Croissance_%'].mean()

print(f"üí∞ PIB Total (Top 20): {pib_total:,.2f} Milliards USD")
print(f"üìà Croissance Moyenne: {croissance_moyenne:.2f}%")
print(f"üåç Nombre de pays analys√©s: {len(df)}")
print()

# Classement Top 10
print("=" * 80)
print(" üèÜ TOP 10 DES √âCONOMIES MONDIALES ".center(80, "="))
print("=" * 80)
print()
print(df[['Rang', 'Pays', 'PIB_Mds_USD', 'Croissance_%', 'PIB_Habitant_USD', 'Region']].head(10).to_string(index=False))
print()

# Analyses sp√©cifiques
print("=" * 80)
print(" üîç ANALYSES CL√âS ".center(80, "="))
print("=" * 80)
print()

# Pays avec meilleure croissance
meilleure_croissance = df.loc[df['Croissance_%'].idxmax()]
print(f"üöÄ Meilleure Croissance: {meilleure_croissance['Pays']} (+{meilleure_croissance['Croissance_%']}%)")

# Pays avec pire croissance
pire_croissance = df.loc[df['Croissance_%'].idxmin()]
print(f"üìâ Croissance la plus faible: {pire_croissance['Pays']} ({pire_croissance['Croissance_%']}%)")

# Pays avec meilleur PIB/habitant
meilleur_pib_hab = df.loc[df['PIB_Habitant_USD'].idxmax()]
print(f"üíé Meilleur PIB/Habitant: {meilleur_pib_hab['Pays']} (${meilleur_pib_hab['PIB_Habitant_USD']:,})")

# √âcart USA-Chine
ecart = df.loc[0, 'PIB_Mds_USD'] - df.loc[1, 'PIB_Mds_USD']
print(f"üá∫üá∏üá®üá≥ √âcart USA-Chine: {ecart:,.2f} Mds USD ({ecart/df.loc[1, 'PIB_Mds_USD']*100:.1f}%)")
print()

# Statistiques par r√©gion
print("=" * 80)
print(" üåç STATISTIQUES PAR R√âGION ".center(80, "="))
print("=" * 80)
print()
print(df_regions.to_string(index=False))
print()

# ============================================
# 3. VISUALISATIONS
# ============================================

fig = plt.figure(figsize=(20, 12))
fig.suptitle('Analyse Compl√®te du PIB Mondial 2024', fontsize=24, fontweight='bold', y=0.995)

# Graphique 1: Classement Top 20 (Barres horizontales)
ax1 = plt.subplot(2, 3, 1)
colors = ['#FF6B6B' if x < 0 else '#4ECDC4' for x in df['Croissance_%']]
bars = ax1.barh(df['Pays'], df['PIB_Mds_USD'], color=colors, alpha=0.8, edgecolor='black')
ax1.set_xlabel('PIB (Milliards USD)', fontsize=12, fontweight='bold')
ax1.set_title('üèÜ Classement Top 20 par PIB Nominal', fontsize=14, fontweight='bold', pad=20)
ax1.invert_yaxis()
ax1.grid(axis='x', alpha=0.3)
# Ajouter les valeurs
for i, (v, pays) in enumerate(zip(df['PIB_Mds_USD'], df['Pays'])):
    ax1.text(v + 500, i, f'{v:,.0f}', va='center', fontsize=8)

# Graphique 2: Taux de croissance
ax2 = plt.subplot(2, 3, 2)
colors_croissance = ['#2ECC71' if x > 2.5 else '#E74C3C' if x < 0 else '#F39C12' for x in df['Croissance_%']]
bars2 = ax2.bar(range(len(df)), df['Croissance_%'], color=colors_croissance, alpha=0.8, edgecolor='black')
ax2.axhline(y=0, color='black', linestyle='-', linewidth=1)
ax2.axhline(y=2.8, color='blue', linestyle='--', linewidth=1, alpha=0.5, label='Moyenne (2.8%)')
ax2.set_xlabel('Pays (par rang)', fontsize=12, fontweight='bold')
ax2.set_ylabel('Croissance (%)', fontsize=12, fontweight='bold')
ax2.set_title('üìà Taux de Croissance 2024', fontsize=14, fontweight='bold', pad=20)
ax2.set_xticks(range(len(df)))
ax2.set_xticklabels(range(1, len(df)+1), fontsize=8)
ax2.grid(axis='y', alpha=0.3)
ax2.legend()

# Graphique 3: PIB par habitant Top 10
ax3 = plt.subplot(2, 3, 3)
top10 = df.head(10).sort_values('PIB_Habitant_USD', ascending=True)
bars3 = ax3.barh(top10['Pays'], top10['PIB_Habitant_USD'], color='#9B59B6', alpha=0.8, edgecolor='black')
ax3.set_xlabel('PIB/Habitant (USD)', fontsize=12, fontweight='bold')
ax3.set_title('üí∞ PIB par Habitant - Top 10', fontsize=14, fontweight='bold', pad=20)
ax3.grid(axis='x', alpha=0.3)
for i, (v, pays) in enumerate(zip(top10['PIB_Habitant_USD'], top10['Pays'])):
    ax3.text(v + 1500, i, f'${v:,}', va='center', fontsize=9)

# Graphique 4: R√©partition par r√©gion (Camembert)
ax4 = plt.subplot(2, 3, 4)
colors_pie = ['#FF6B6B', '#4ECDC4', '#95E1D3', '#F38181', '#AA96DA', '#FCBAD3']
wedges, texts, autotexts = ax4.pie(df_regions['PIB_Mds_USD'],
                                     labels=df_regions['Region'],
                                     autopct='%1.1f%%',
                                     colors=colors_pie,
                                     startangle=90,
                                     textprops={'fontsize': 10, 'weight': 'bold'})
ax4.set_title('üåç R√©partition du PIB par R√©gion', fontsize=14, fontweight='bold', pad=20)

# Graphique 5: √âvolution USA vs Chine vs UE (2020-2024)
ax5 = plt.subplot(2, 3, 5)
ax5.plot(df_evolution['Annee'], df_evolution['USA'], marker='o', linewidth=3,
         label='√âtats-Unis', color='#3498DB', markersize=10)
ax5.plot(df_evolution['Annee'], df_evolution['Chine'], marker='s', linewidth=3,
         label='Chine', color='#E74C3C', markersize=10)
ax5.plot(df_evolution['Annee'], df_evolution['UE'], marker='^', linewidth=3,
         label='Union Europ√©enne', color='#2ECC71', markersize=10)
ax5.set_xlabel('Ann√©e', fontsize=12, fontweight='bold')
ax5.set_ylabel('PIB (Milliards USD)', fontsize=12, fontweight='bold')
ax5.set_title('üìä √âvolution 2020-2024: Grandes Puissances', fontsize=14, fontweight='bold', pad=20)
ax5.legend(fontsize=10, loc='upper left')
ax5.grid(True, alpha=0.3)
ax5.set_xticks(df_evolution['Annee'])

# Graphique 6: Heatmap Croissance vs PIB/Habitant
ax6 = plt.subplot(2, 3, 6)
scatter = ax6.scatter(df['PIB_Habitant_USD'], df['Croissance_%'],
                      s=df['PIB_Mds_USD']/50, # Taille = PIB
                      c=df['PIB_Mds_USD'],
                      cmap='viridis',
                      alpha=0.6,
                      edgecolors='black',
                      linewidth=1.5)
# Ajouter les noms des pays importants
for i, row in df.head(10).iterrows():
    ax6.annotate(row['Pays'],
                (row['PIB_Habitant_USD'], row['Croissance_%']),
                fontsize=8,
                alpha=0.7,
                xytext=(5, 5),
                textcoords='offset points')
ax6.set_xlabel('PIB par Habitant (USD)', fontsize=12, fontweight='bold')
ax6.set_ylabel('Croissance (%)', fontsize=12, fontweight='bold')
ax6.set_title('üéØ PIB/Habitant vs Croissance\n(Taille = PIB total)', fontsize=14, fontweight='bold', pad=20)
ax6.axhline(y=0, color='red', linestyle='--', linewidth=1, alpha=0.5)
ax6.grid(True, alpha=0.3)
plt.colorbar(scatter, ax=ax6, label='PIB (Mds USD)')

plt.tight_layout()
plt.savefig('analyse_pib_2024.png', dpi=300, bbox_inches='tight')
print("‚úÖ Graphique sauvegard√©: 'analyse_pib_2024.png'")
plt.show()

# ============================================
# 4. EXPORT DES DONN√âES
# ============================================

# Export CSV
df.to_csv('pib_mondial_2024.csv', index=False, encoding='utf-8-sig')
print("‚úÖ Donn√©es export√©es: 'pib_mondial_2024.csv'")

# Export Excel avec formatage
with pd.ExcelWriter('analyse_pib_2024.xlsx', engine='openpyxl') as writer:
    df.to_excel(writer, sheet_name='Classement_Complet', index=False)
    df_regions.to_excel(writer, sheet_name='Par_Region', index=False)
    df_evolution.to_excel(writer, sheet_name='Evolution_2020-2024', index=False)
print("‚úÖ Rapport Excel cr√©√©: 'analyse_pib_2024.xlsx'")

# ============================================
# 5. INSIGHTS ET CONCLUSIONS
# ============================================

print()
print("=" * 80)
print(" üí° INSIGHTS CL√âS 2024 ".center(80, "="))
print("=" * 80)
print()

insights = [
    "1. üá∫üá∏ Les √âtats-Unis dominent avec 29 168 Mds $, soit 28% du PIB mondial",
    "2. üáÆüá≥ L'Inde affiche la meilleure croissance (+7%) et pourrait devenir 3e d'ici 2026",
    "3. üáØüáµ Le Japon est le seul pays du Top 10 en r√©cession (-0.5%)",
    "4. üá™üá∏ L'Espagne brille en Europe avec +5.5% de croissance",
    "5. üåè L'Asie repr√©sente 30.2% du PIB mondial contre 32.5% pour l'Am√©rique du Nord",
    "6. üìâ L'√©cart USA-Chine reste important: +59.6% en faveur des √âtats-Unis",
    "7. üáµüá± La Pologne affiche une croissance robuste de +4.9%",
    "8. üí∞ L'√©cart de PIB/habitant est √©norme: de $2 730 (Inde) √† $86 530 (USA)"
]

for insight in insights:
    print(f"  {insight}")
print()
print("=" * 80)
print(" üé¨ Analyse termin√©e avec succ√®s! ".center(80, "="))
print("=" * 80)
```
